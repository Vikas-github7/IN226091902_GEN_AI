# Task 1: Conceptual Understanding 

**1. What is the difference between "Love" and "love" in NLP?**

To a computer, "Love" (with a capital L) and "love" are treated as two completely distinct tokens because their underlying ASCII/Unicode characters differ. If we do not lowercase the text, a machine learning model will learn separate weights and representations for "Love" and "love", splitting the frequency of the word and reducing the model's efficiency.


**2. What happens if stopwords are not removed?**

If stopwords (common words like "is", "the", "and") are not removed, they will dominate the frequency distribution of your text corpus. This increases the vocabulary size, consumes more memory and processing power, and can add noise that distracts the algorithm from the truly meaningful, context-specific words.

**3. Provide two real-world scenarios where removing stopwords can be harmful.**
- Sentiment Analysis: Removing words like "not" or "no" can completely flip the meaning of a sentence  (e.g., "The movie was not good" becomes "movie good").
- Machine Translation or Text Generation: These tasks require an understanding of grammar, syntax, and phrasing to produce human-readable sentences. Removing stopwords destroys the grammatical structure.

**4. Explain the difference between stemming and lemmatization.**
- Stemming is a crude, rule-based heuristic that simply chops off prefixes or suffixes from a word (e.g., "running" becomes "run", but "better" might stay "better", and "caring" might become "car"). It is fast but often results in non-words.
- Lemmatization is a more advanced technique that uses morphological analysis and a dictionary (vocabulary) to reduce a word to its linguistic base form, known as a lemma (e.g., "better" becomes "good", "running" becomes "run"). It is slower but contextually accurate.

# Task 2:Build Advanced Preprocessing Function

```python
import re
from collections import Counter
def preprocess_text(text):
    """
    Transforms raw text into clean, structured tokens.
    Handles noise, normalization, and filters short words.
    """
    # Error Handling for non-string inputs or empty strings
    if not isinstance(text, str) or not text.strip():
        return [], ""
    
    # 1. Convert text to lowercase
    text = text.lower()
    
    # 2. Remove URLs and email-like patterns
    text = re.sub(r'https?://\S+|www\.\S+|[\w\.-]+@[\w\.-]+', '', text)
    
    # 3. Remove numbers
    text = re.sub(r'\d+', '', text)
    
    # 4. Remove punctuation and emojis (keep only alphabetic characters and spaces)
    text = re.sub(r'[^a-z\s]', ' ', text)
    
    # 5. Handle repeated characters (e.g., "soooo goooood" -> "soo good")
    # We replace 3 or more consecutive characters with 2 to preserve words like "good"
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)
    
    # 6. Tokenize (this automatically handles removing extra spaces)
    tokens = text.split()
    
    # 7. Remove very short tokens (length <= 2), keeping meaningful exceptions
    exceptions = {"no", "not"}
    clean_tokens = [word for word in tokens if len(word) > 2 or word in exceptions]
    
    # 8. Reconstruct the clean sentence
    clean_sentence = " ".join(clean_tokens)
    
    return clean_tokens, clean_sentence
```

```python
list_sentence = [ "I have 2 dogs",  "This is good",  "soooo goooood!!!", "WOW!!! This is GREAT!!!" , "Visit http://example.com now"]

for i in list_sentence:
    print("Before Processing: ",i)
    print("After Processing: ")
    tokens,sentence = preprocess_text(i)
    print("Tokens:", tokens)
    print("Cleaned sentence:", sentence)
    print("--"*30)
```

# Task 3: Stress Testing


```python
sample_input = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

processed_data = []

for i in sample_input:
    print("Original Text: ", i)
    tokens, sentence = preprocess_text(i)

    # Storing for later tasks
    processed_data.append({
        "original": i,
        "tokens": tokens,
        "clean_sentence": sentence
    })
    
    print("Cleaned Tokens: ", tokens)
    print("Cleaned Sentence: ", sentence)
    print("--"*30)
    
```

# Task 4: Token Analytics

```python

for data in processed_data:
    tokens = data["tokens"]
    original = data["original"]
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    
    # Calculate average length, handling division by zero for empty token lists
    if total_tokens > 0:
        avg_length = sum(len(word) for word in tokens) / total_tokens
    else:
        avg_length = 0.0
        
    print(f"Sentence: {original}:")
    print(f"  Total Tokens : {total_tokens}")
    print(f"  Unique Tokens: {unique_tokens}")
    print(f"  Avg Length   : {avg_length:.2f}")
    print("--" * 30)
```

# Task 5: Frequency Analysis


```python
from collections import Counter

# Combine all tokens from all sentences
all_tokens = []
for data in processed_data:
    all_tokens.extend(data["tokens"])

# Calculate frequencies
token_counts = Counter(all_tokens)

print("Top 10 most frequent words:")
for word, count in token_counts.most_common(10):
    print(f"  {word}: {count}")

print("\nTop 5 least frequent words:")
# Reverse the most_common list to get the least frequent
for word, count in reversed(token_counts.most_common()[-5:]):
    print(f"  {word}: {count}")
```

# Task 6: Build Full Pipeline

```python
def full_pipeline(text_list):
    pipeline_output = {
        "tokens": [],
        "clean_sentences": []
    }
    
    for text in text_list:
        tokens, clean_sentence = preprocess_text(text)
        pipeline_output["tokens"].append(tokens)
        pipeline_output["clean_sentences"].append(clean_sentence)
        
    return pipeline_output

# Quick test of the pipeline
pipeline_results = full_pipeline(sample_input[:2]) # Testing just the first 2
print(pipeline_results)
```

# Task 7: Error Handling


```python
error_test_cases = [
    "",             # Empty string
    "😂😂😂",       # Only emojis
    "12345 67890"   # Only numbers
]

print("ERROR HANDLING")
error_results = full_pipeline(error_test_cases)

for i in range(len(error_test_cases)):
    print(f"Original: '{error_test_cases[i]}'")
    print(f"Tokens  : {error_results['tokens'][i]}")
    print(f"Sentence: '{error_results['clean_sentences'][i]}'")
    print("-" * 30)
```
